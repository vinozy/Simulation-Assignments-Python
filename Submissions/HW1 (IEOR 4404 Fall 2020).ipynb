## Setting

**Negative Binomial Distribution** is a discrete probability distribution that models the number of successes (Our outcome variable: $k$) in a sequence of independent and identically distributed Bernoulli trials (with the probability of success=$p$) before a specified (non-random) number of failures (denoted $r$) occurs (Source: Wikipedia [Negative Binomial Distribution](https://en.wikipedia.org/wiki/Negative_binomial_distribution)).

The probability mass function follows

$P(k)=\frac{(k+r-1)!}{k!(r-1)!}p^k(1-p)^r$ where $k=0, 1, 2, ..., \infty$


**An application of Negative Binomial Distribution:**

Pokemon Go is a mobile app game in which users encounter different pokemon one by one and catch them. Shiny pokemon is a special version of pokemon, which has a different color from the normal version.

##Q6. [10 points]

* Use inverse transform method to generate 1000 samples. Start the search from $k=0$.
* Plot both the theoretical distribution and the sample distribution using the same plot

**[Since $x$ can reach $\infty$, remember to use a while loop.]**

**[You are not allowed to use scipy.stats.nbinom to get the PMF or CDF]**


```
def nbinomial_sample():
    u=np.random.rand()   # random number b/w 0 and 1
    p=0.9     # probability of a non-shiny pokemon occurs
    r=3        # number of shiny pokemons occurred
    x=0       # starting point for searching
    pmf=spsp.comb(x+r-1, x)*(p**x)*((1-p)**r)
    cdf=pmf    # initialized CDF

    while u>=cdf:
        # update x
        x+=1
        # update PMF and CDF
        pmf=spsp.comb(x+r-1, x)*(p**x)*((1-p)**r)
        cdf+=pmf
    return x
```

```
# generate 1000 samples using list comprehension
samples=[nbinomial_sample() for i in range(1000)]
samples
```

```
# plot both the theoretical and sample distribution using the same plot
p=0.9
r=3
values,counts=np.unique(np.array(samples),return_counts=True)
plt.scatter(values,counts/np.sum(counts),label="sample",zorder=2)
x=np.arange(np.max(values)+1)
pmf=spsp.comb(x+r-1, x)*(p**x)*((1-p)**r)
plt.bar(x,pmf,label="theory")
plt.legend(loc="best")
plt.xlabel("outcomes")
plt.ylabel("probabilities")
plt.show()
```

## Q7. [10 points]
* Use inverse transform method to generate 1000 samples. Perform an upward/downward search from $\lfloor E(k) \rfloor$.
* Plot both the theretical distribution and the sample distribution using the same plot.

```
def updown_nbinomial_sample():
    p=0.9     # probability of a non-shiny pokemon occurs
    r=3        # number of shiny pokemons occurred
    es=(r*p)/(1-p)    # compute E(k)

    # (1) decide the starting point, x, for search
    x = int(es)   # or x = np.floor(lmbda)      # let x = 3

    # (2) compute the CDF corresponds to the starting point
    i = 0
    cdf=0
    while i<=x:
        pmf=spsp.comb(i+r-1, i)*(p**i)*((1-p)**r)   
        cdf+=pmf
        i+=1
    # got cdf(3) after the loop

    # (3) generate a random number u, compare the random number u with the CDF of the starting point x
    u = np.random.rand()

    # (4) if u >= CDF(x): search upward
    if u>=cdf:
        x+=1  # x=3+1=4
        pmf=spsp.comb(x+r-1, x)*(p**x)*((1-p)**r) # pmf(4)
        cdf+=pmf # cdf(4) = cdf(3)+pmf(4)
        while u>=cdf:
            x+=1 # x=4+1=5
            pmf=spsp.comb(x+r-1, x)*(p**x)*((1-p)**r) # pmf(5)
            cdf+=pmf # cdf(5) = cdf(4)+pmf(5)
        return x

    # (5) else: search downward
    else:
        cdf-=pmf     # cdf(2) = cdf(3)-pmf(3)
        while u<cdf:
            x-=1 # x=2
            pmf=spsp.comb(x+r-1, x)*(p**x)*((1-p)**r) # pmf(2)
            cdf-=pmf # cdf(1) = cdf(2)-pmf(2)
        return x
```

```
# generate 1000 samples using list comprehension
samples=[updown_nbinomial_sample() for i in range(1000)]
samples
```

```
# plot both the theoretical and sample distribution using the same plot
p=0.9
r=3
values,counts=np.unique(np.array(samples),return_counts=True)
plt.scatter(values,counts/np.sum(counts),label="sample",zorder=2)
x=np.arange(np.max(values)+1)
pmf=spsp.comb(x+r-1, x)*(p**x)*((1-p)**r)
plt.bar(x,pmf,label="theory")
plt.legend(loc="best")
plt.xlabel("outcomes")
plt.ylabel("probabilities")
plt.show()
```

##Setting

Benford's law, also called the first-digit law, is an observation of the frequency distribution of leading digits in many real-life sets of numerical data. The law states that in many naturally occurring collections of numbers, the leading significant digit is likely to be small. For example, in sets that obey the law, the number 1 appears as the most significant digit about 30% of the time, while 9 appears as the most significant digit less than 5% of the time. Benford's law also makes predictions about the distribution of second digits, third digits, digit combinations, and so on.

It has been shown that this result applies to a wide variety of data sets, including electricity bills, stock prices, house prices, income tax, corporate disbursements, sales figures,  demographics data, and scientific data.

A number is said to satisfy Benford's law if the leading digit d (d ∈ {1, ..., 9}) occurs with probability :

$P(d)=\log _{10}(d+1)-\log _{10}(d)=\log _{10}\left({\frac {d+1}{d}}\right)=\log _{10}\left(1+{\frac {1}{d}}\right)$

```
d=sympy.Symbol("d",positive=True)
```

```
# first, derive the CDF in the closed form formula
PMF = sympy.log(1+1/d, 10)
PMF
```

```
CDF = sympy.Sum(PMF,(d,1,d)).doit().simplify()
CDF
```

```
# because the log of a product is the sum of the logs: log(x*y) = log(x) +log(y)
# transform the numerator into the log of the PRODUCT of log(1+1/d), where d is from 1 to d.
CDF = sympy.log(    sympy.Product(1+1/d,(d,1,d)).doit().simplify()    )    /    sympy.log(10)
CDF

# therefore, CDF = log(d+1)/log(10)
```

```
# solve u<CDF(d)
u=sympy.Symbol("u")
S1=sympy.solve(CDF-u,d)
S1=S1[0].collect(u)
S1
```

```
# solve CDF(d)<=u
S2=sympy.solve(CDF.subs({d:d-1})-u,d)
S2=S2[0].collect(u)
S2
```

```
# therefore, the direct conversion formula is:
d = sympy.floor(S2)
d
```

```
# generate 10000 samples using the direct conversion formula above
u=np.random.rand(10000)
samples=np.floor(np.exp(u*np.log(10)))
samples
```
